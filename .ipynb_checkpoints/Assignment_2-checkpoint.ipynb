```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as scp
%matplotlib inline
```

```python
# Generating 1000 random numbers with a mean of 0 and a variance of 1
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
# Generating the standard Rayleigh distribution
z = np.sqrt(x**2 + y**2)



# Plotting the histogram in 50 bins and normalizing
plt.hist(z, bins = 50, normed=True)
plt.title("Histogram of a standard Rayleigh distribution")
plt.xlabel("z")
plt.ylabel("Probability")
plt.show()
```

```python
# Generating the N numbers to be used to calculate the sample mean, variance and skewness
n = np.linspace(1, 6, 6)
N = 10**n

# Create empty lists for the mean, variance and skewness
mean_list = []
var_list = []
skew_list = []

# Define functions for the mean, variance and skewness
def mean(x):
    m = np.sum(x)/len(x)
    return m

def var(m, x):
    v = np.sum(x**2)/len(x) - m**2
    return v

def skew(m, x):
    s = (1/len(x) * np.sum((x-m)**3)) / (((1/len(x) * np.sum((x-m)**2)))**(3/2))
    return s

# Calculate the mean, variance and skewness given our Rayleigh distribution
for i in range(len(N)):
    x = np.random.normal(0, 1, N[i])
    y = np.random.normal(0, 1, N[i])
    z = np.sqrt(x**2 + y**2)

    mean_list.append(mean(z))
    var_list.append(var(mean_list[i], z))
    skew_list.append(skew(mean_list[i], z))
    

# Plotting the mean, variance and skewness as function of N

plt.subplot(221)
plt.semilogx(N, mean_list)
plt.title("Mean as function of N")
plt.ylabel("Mean")
plt.xlabel("N")
plt.subplot(222)
plt.semilogx(N, var_list)
plt.title("Variance as function of N")
plt.ylabel("Variance")
plt.xlabel("N")
plt.subplot(223)
plt.semilogx(N, skew_list)
plt.title("Skewness as function of N")
plt.ylabel("Skewness")
plt.xlabel("N")
plt.tight_layout()
plt.show()
```

```python
# Analytic solution of the fourier transform
def analytic_fft(x):
    y = 2*np.sinc(x)
    return y

# Creating our x space
x = np.linspace(-5, 5, 1000)


f = []

# Generate the value of 1 between -1 and 1
for i in range(len(x)):
    if -1 <= x[i] <= 1:
        f.append(1)
    else:
        f.append(0)
        

# In order to fft properly we need to shift the fft 
f = np.fft.fftshift(f) 
f = np.fft.fft(f)
f = np.fft.fftshift(f)
x = np.fft.fftfreq(len(x), 10/1000)
x = np.fft.fftshift(x)
ax = plt.subplot(111)
ax.set_xlim([-4,4])
ax.plot(x,(50*analytic_fft(x)), label="Analytic")
ax.plot(x,f, label = "FFT")
plt.title("Fourier transform comparison")
ax.legend()
plt.show()
```

```python
# Generating x space
x = np.linspace(0, 10, 100)

# Creating empty list y
y = []

# "Sketching" the non-zero region
for i in range(len(x)):
    if x[i] <=1:
        y.append(x[i])
    else:
        y.append(1)

ax = plt.subplot(111)
ax.plot(x,y)
ax.set_xlim([-1, 4])
ax.set_ylim([-1 ,4])
ax.fill_between(x, 0, y)
plt.title("Sketch of non-zero region")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
